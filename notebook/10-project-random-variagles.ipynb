```python
from scipy.stats import binom, norm, uniform
import numpy as np
import matplotlib.pyplot as plt


np.random.seed(42)
```

## Binomial distribution

En teoría de la probabilidad y estadística, la distribución binomial es la distribución de probabilidad discreta que da sólo dos resultados posibles en un experimento; por ejemplo, al lanzar una moneda, puede salir cara o cruz.

```python
def dbinom(x, size, prob=0.5):
    """
    Calculates the point estimate of the binomial distribution
    """
    result = binom.pmf(k=x, n=size, p=prob, loc=0)
    return result


def pbinom(q, size, prob=0.5):
    """
    Calculates the cumulative of the binomial distribution
    """
    result = binom.cdf(k=q, n=size, p=prob, loc=0)
    return result


def qbinom(p, size, prob=0.5):
    """
    Calculates the quantile function from the binomial distribution
    """
    result = binom.ppf(q=p, n=size, p=prob, loc=0)
    return result


def rbinom(n, size, prob=0.5):
    """
    Generates random variables from the binomial distribution
    """
    result = binom.rvs(n=size, p=prob, size=n)
    return result
```

```python
data = rbinom(2000, size=10, prob=0.2)
pmf = dbinom(data, size=10, prob=0.2)
cdf = pbinom(data, size=10, prob=0.2)
ppf = qbinom(data, size=10, prob=0.2)
```

```python
fig, axis = plt.subplots(3, 1, figsize=(5, 7))
axis[0].stem(data, pmf, "r-")
axis[1].stem(data, cdf, "b-")
axis[2].stem(data, ppf, "g-")

plt.tight_layout()  # Adjust the layout
plt.show()  # Show the plot
```

### Uniform distribution

```python
np.random.seed(42)
```

```python
def point_est_unif(x, low=0, high=1):
    result = uniform.pdf(x, loc=low, scale=(high-low))
    return result
low = 0
high = 1
data = np.arange(-5, 5, 0.01)
pmf = point_est_unif(data, low=low, high=high)
pmf
```

```python

def cumulative_unif(q, low =0, high=1):
    result = uniform.cdf(q, loc=low, scale=(high-low))
    return result
low = 0
high = 1
data = np.arange(-5, 5, 0.01)
cdf = cumulative_unif(data, low=low, high=high)
cdf
```

```python
def quantile_unif(p, low=0, high=1):
    result = uniform.ppf(p, loc=low, scale=(high-low))
    return result
low = 0
high = 1
data = np.arange(-5, 5, 0.01)
ppf = quantile_unif(data, low=low, high=high)
ppf
```

```python

def random_unif(n, low=0, high=1):
    result = uniform.rvs(loc=low, scale=(high-low), size=n)
    return result
low = 0
high = 1
data = np.arange(-5, 5, 0.01)
n = 10
random_samples = random_unif(n=n, low=low, high=high)
random_samples
```

```python
fig, axis = plt.subplots(3, 1, figsize=(5, 7))

axis[0].plot(data, pmf, "r-")
axis[1].plot(data, cdf, "b-")
axis[2].plot(data, ppf, "g-")

plt.tight_layout()
plt.show()
```
